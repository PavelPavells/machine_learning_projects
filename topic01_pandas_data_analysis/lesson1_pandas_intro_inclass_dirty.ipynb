**Логическая индексация**

```python
# Сравнение заработка(wage) между замужними и свободными
print('Женщины: ', data[data['female'] == 1]['wage'].mean())
print('Мужчины: ', data[data['female'] == 0]['wage'].mean())
#data[data['female'] == 1]['wage'].mean(), \
#data[data['female'] == 0]['wage'].mean()
```

```python
#Сравнение женщин (Замужних и незамужних)
print('Замужние: ', data[(data['female'] == 0) & (data['married'] == 1)]['wage'].median())
print('Свободные: ', data[(data['female'] == 0) & (data['married'] == 0)]['wage'].median())
#data[(data['female'] == 0) & (data['married'] == 1)]['wage'].median(), \
#data[(data['female'] == 0) & (data['married'] == 0)]['wage'].median()
```
